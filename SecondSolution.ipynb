<h1>Feature Engineering</h1>

```python
#Processing
def get_combined_dataframe():
    train = pd.read_csv('train.csv')
    test = pd.read_csv('test.csv')
    targets = train.Survived
    train.drop('Survived',1,inplace=True)
    combined = train.append(test)
    combined.reset_index(inplace=True)
    combined.drop('index',1,inplace=True)
    return combined,targets
combined,targets = get_combined_dataframe()
combined.count()

def get_titles():
    global combined
    combined['Title'] = combined['Name'].map(lambda name: name.split(',')[1].split('.')[0].strip())
    # a map of more aggregated titles
    Title_Dictionary = {
                        "Capt":       "Officer",
                        "Col":        "Officer",
                        "Major":      "Officer",
                        "Jonkheer":   "Royalty",
                        "Don":        "Royalty",
                        "Sir" :       "Royalty",
                        "Dr":         "Officer",
                        "Rev":        "Officer",
                        "the Countess":"Royalty",
                        "Dona":       "Royalty",
                        "Mme":        "Mrs",
                        "Mlle":       "Miss",
                        "Ms":         "Mrs",
                        "Mr" :        "Mr",
                        "Mrs" :       "Mrs",
                        "Miss" :      "Miss",
                        "Master" :    "Master",
                        "Lady" :      "Royalty"

                        }
    combined['Title'] = combined.Title.map(Title_Dictionary)
get_titles()
combined.head()
```

```python
#Processing age
grouped_train = combined.head(891).groupby(['Sex','Pclass','Title'])
grouped_median_train = grouped_train.median()
print grouped_median_train

combined["Age"] = combined.groupby(['Sex','Pclass','Title'])['Age'].transform(lambda x: x.fillna(x.median()))
combined['Age'] = (combined.Age - combined.Age.min())/(float)(combined.Age.max() - combined.Age.min()) 
```

```python
#Processing titles
def process_name():
    global combined
    combined.drop('Name',axis=1,inplace=True)
    dummy_features = pd.get_dummies(combined['Title'],prefix='Title')
    combined = pd.concat([combined,dummy_features],axis=1)
    combined.drop('Title',axis=1,inplace=True)
process_name()
combined.head()
```

```python
#Processing fare
#Missing a single value
def fill_fare():
    global combined
    combined.head(891).Fare.fillna(combined.head(891).Fare.mean(),inplace=True)
    combined.iloc[891:].Fare.fillna(combined.iloc[891:].Fare.mean(),inplace=True)
fill_fare()
combined.Fare = (combined.Fare - combined.Fare.min())/(combined.Fare.max() - combined.Fare.min())
combined.count()
```

```python
def fill_embarked():
    global combined
    combined.head(891).Embarked.fillna(combined.head(891).Embarked.mode()[0],inplace=True)
    combined.iloc[891:].Embarked.fillna(combined.iloc[891:].Embarked.mode()[0],inplace=True)
    
    dummy_row = pd.get_dummies(combined.Embarked,prefix='Embarked')
    combined = pd.concat([combined,dummy_row],axis=1)
    combined.drop('Embarked',axis=1,inplace=True)
fill_embarked()
combined.head()
```

```python
#Analysis for cabin
def filter_cabin():
    global combined
    combined.Cabin.fillna('U',inplace=True)
    combined.Cabin = combined.Cabin.map(lambda cabin : cabin[0])
    dummy = pd.get_dummies(combined.Cabin,prefix='Cabin')
    combined = pd.concat([combined,dummy],axis=1)
    combined.drop('Cabin',axis=1,inplace=True)
filter_cabin()
combined.head()
```

```python
#Processing Sex
def process_sex():
    global combined
    combined.Sex = map(lambda x : 1 if x == 'male' else 0,combined.Sex)
process_sex()
combined.head()
```

```python
#Processing pclass
def process_pclass():
    global combined
    dummy = pd.get_dummies(combined.Pclass,prefix='Pclass')
    combined = pd.concat([combined,dummy],axis=1)
    combined.drop('Pclass',axis=1,inplace=True)
process_pclass()
combined.head()
```

```python
#Process tickets
#Little bit strange
def process_ticket():
    global combined
    
    def clean_ticket(ticket):
        ticket = ticket.replace('.','')
        ticket = ticket.replace('/','')
        ticket = ticket.split()
        ticket = map(lambda t: t.strip(),ticket)
        ticket = filter(lambda t: not t.isdigit(),ticket)
        if len(ticket) > 0:
            return ticket[0]
        else:
            return 'XXX'
#     combined.Ticket = combined.Ticket.map(clean_ticket)
#     dummy = pd.get_dummies(combined['Ticket'],prefix='Ticket')
#     combined = pd.concat([combined,dummy],axis=1)
    combined.drop('Ticket',axis=1,inplace=True)
process_ticket()
combined.head()
```

```python
#Processing Parch and SibSp
def process_family():
    global combined
    combined['FamilySize'] = combined['SibSp'] + combined['Parch']
    combined['Singleton'] = combined.FamilySize.map(lambda x : 1 if x == 1 else 0)
    combined['SmallFamily'] = combined.FamilySize.map(lambda x : 1 if 2<=x<=4 else 0)
    combined['LargeFamily'] = combined.FamilySize.map(lambda x : 1 if x == 1 else 0)    
    combined.drop('FamilySize',axis=1,inplace=True)
    combined.drop('SibSp',axis=1,inplace=True)
    combined.drop('Parch',axis=1,inplace=True)
process_family()
combined.head()
```
